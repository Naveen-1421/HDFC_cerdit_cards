```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
#Inspect the website and webscraping
```

```python
#to scrap the website
url = "https://www.hdfcbank.com/personal/pay/cards/credit-cards"
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
data = soup.find_all('div', {'class': 'cardparent'})
#to create the loop for the columns
card_name = []
card_fees = []
reward_points = []
lounge_access = []
milestone_benefits = []
fee_reversals = []
# append for each columns
for card in data:
    name = card.find('h2', {'class': 'cardTitle'})
    card_fee = card.find('div', {'class':"accordion"})
    reward = card.find('div', {'class': 'offer-dtl mB10'})
    lounge = card.find('span', {'class': 'card-cat'})
    milestone = card.find('div', {'class': 'col-sm-12 mTn10 card-des'})
    fee_reversal = card.find('div', {'class': 'be-ex-content'})
    card_name.append(name.text)
    card_fees.append(card_fee)
    reward_points.append(reward.text)
    lounge_access.append(lounge) 
    milestone_benefits.append(milestone.text)
    fee_reversals.append(fee_reversal)
#to create the data frame
data = pd.DataFrame({
    'Card_Name': card_name,
    'Card_Fee': card_fees,
    'Reward_Points_Percentage_per_100_Spent': reward_points,
    'Lounge_Access': lounge_access,
    'Milestone_Benefit': milestone_benefits,
    'fee_reversals': fee_reversals})
data
```

```python
# to relace values
data['Reward_Points_Percentage_per_100_Spent']=data['Reward_Points_Percentage_per_100_Spent'].str.replace('\n', '').str.strip()
data
```

```python
#the data convert the excel file
data.to_excel('HDFC_CARD.xlsx',index=False)
```
